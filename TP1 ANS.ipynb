# I. Réduction de dimensions et Visualisation des données

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

data = pd.read_csv('C:/Users/ordi/Desktop/Apprentissage_Auto/villes.csv', sep=';')
X = data.iloc[:, 1:13].values
labels = data.iloc[:, 0].values

scaler = StandardScaler()
X_std = scaler.fit_transform(X)
pca = PCA()
X_pca = pca.fit_transform(X_std)
components = pca.components_

explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)
n_components = np.argmax(cumulative_variance >= 0.7) + 1
print(f"Nombre d'axes pour conserver 70% de l'information : {n_components}")

noms_variables = data.columns[1:13] 

print("Interprétation de l'Axe 1:")
for i, variable in enumerate(noms_variables):
    print(f"{variable}: {components[0, i]}")

print("\n Interprétation de l'Axe 2:")
for i, variable in enumerate(noms_variables):
    print(f"{variable}: {components[0, i]}")


plt.scatter(X_pca[:, 0], X_pca[:, 1])

for label, x, y in zip(labels, X_pca[:, 0], X_pca[:, 1]):
    plt.annotate(label, xy=(x, y), xytext=(-0.2, 0.2), textcoords='offset points')

plt.xlabel('Premier Axe Principal')
plt.ylabel('Deuxième Axe Principal')
plt.title('Projection des villes dans le premier plan principal')

plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 

data = pd.read_csv('C:/Users/ordi/Desktop/Apprentissage_Auto/villes.csv', sep=';')
X = data.iloc[:, 1:13].values
labels = data.iloc[:, 0].values

scaler = StandardScaler()
X_std = scaler.fit_transform(X)

pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_std)

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X_std)
cluster_labels = kmeans.labels_

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

colors = ['red','yellow','blue','pink']
for i in range(3):
    ax.scatter(X_pca[cluster_labels == i, 0], X_pca[cluster_labels == i, 1], X_pca[cluster_labels == i, 2], c=colors[i], label=f'Cluster {i}')

for label, x, y, z in zip(data['ville'], X_pca[:, 0], X_pca[:, 1], X_pca[:, 2]):
    ax.text(x, y, z, label, bbox=dict(facecolor='white', alpha=0.5))

ax.set_xlabel('Premier Axe Principal')
ax.set_ylabel('Deuxième Axe Principal')
ax.set_zlabel('Troisième Axe Principal')
plt.title('PCA en 3D avec 3 Clusters')
plt.legend()

plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np

silhouette = []

for i in np.arange(2, 6):
    clustering = KMeans(n_clusters=i, random_state=0).fit_predict(X)
    silhouette_score = metrics.silhouette_score(X, clustering, metric='euclidean')
    silhouette.append(silhouette_score)

num_clus = np.argmax(silhouette) + 2  
max_silhouette= max(silhouette)

print("Nombre optimal de clusters:", num_clus)
print("Indice Silhouette maximal:", max_silhouette)
```
